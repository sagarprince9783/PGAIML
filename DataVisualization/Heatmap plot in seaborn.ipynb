# Heatmap plot in seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=np.linspace(1,10,20).reshape(4,5)
```

```python
var
```

```python
y={'fontsize':20,'color':'r'}
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=10)
plt.show()
```

```python
y={'fontsize':20,'color':'r'}
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=1)
plt.show()
```

```python
y={'fontsize':20,'color':'r'}
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=1,linecolor='y')
plt.show()
```

```python
# cbar is used for the color bar 
y={'fontsize':20,'color':'r'}
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=1,linecolor='y',cbar=False)
plt.show()
```

```python
# cbar is used for the color bar 
y={'fontsize':20,'color':'r'}
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=1,linecolor='y',cbar=False,xticklabels=False)
plt.show()
```

```python
# cbar is used for the color bar 
y={'fontsize':20,'color':'r'}
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=1,linecolor='y',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
sns.heatmap(var,vmin=0,vmax=20,cmap='gist_heat')
plt.show()
```

```python
sns.heatmap(var)
plt.show()
```
